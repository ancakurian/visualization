Cultural Festivals Funding Program (CFFP) Recipients

source: https://open.toronto.ca/dataset/cultural-festivals-funding-program-recipients/

The following visualization uses data provided by the City of Toronto with regards to the cultural events and the local organizations it supports through its grants. The data includes only the 2024 grants information. 

To any intersted parties, be it city counsel, community leader or member of the electorate wanting to know which local organizations are funded with public funds, the graphs will show the grants recievers, the amount awarded and the type of cultural events supported by the City governing body. 

The purpose of the graph is informative and the substantive principle is the main quality. The data is presented so that the beneficiaries are clearly identifiable and the amount obtained in grants is clear to the reader. While the aesthetic aspect is not sought, the two graphs use two different colors (as the histogram bars would be a bold block of red, inhibiting rather than explaining the data). For accessibility, the colors used are red and green. The line graph uses a marker for the $ value awarded. 

To convey how the various cultural events are supported financially, the graph "Cultural Festivals Funding Program Recipients, 2024" presents the data starting with the highest grant awarded at the top and continuing with the smaller grants, in descending order. At a glance, the first three entries on the y axis are the recipients of the highest $ value grants, with the rest on the entries recieving a comparable amount but much smaller than the first three events. 

The second graph, "Cultural Festivals Funding by Program Stream, 2024", shows which type of events are funded by public funds. With the exception of three cultural events, all of them are stand-alone "projects". Even if an event takes place in consecutive years, the grant is awarded annually, with no multi-year financial planning being required from the organization offering the grant (i.e, city of Toronto). 

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python

# Read the CSV file
df = pd.read_csv(r'C:\Users\Anca\Desktop\UofT DSI\visualization\02_activities\assignments\2024 Cultural Festivals Funding Program recipients.csv')

# Display the dataset, first 5 rows and column title
df.head()
```

```python
# Visualize the funding recipients, in ascending order of funding value
fig, ax = plt.subplots(figsize=(15,10))

# Sort by the 'Grant' column in ascending order
df_sorted = df.sort_values(by='Grant', ascending=True)

# Plot the data in a grid, line graph
x = df_sorted["Grant"]
y = df_sorted["Project Name"]

ax.plot(x, y,
        marker='o',
        color='red',
        linewidth=2)

# Reduced the number of characters displayed in the y labels, as many of them has a long, descriptive name 
ax.set_yticks(range(len(y)))
ax.set_yticklabels([label[:30] for label in y])
ax.set_title("Cultural Festivals Funding Program Recipients, 2024",
             fontsize = 20)
ax.set_ylabel("Cultural Event",
              fontsize = 14)
              
ax.set_xlabel("Grant in $",
              fontsize = 14)

ax.grid()
plt.show()
```

```python
# Visualize the type of cultural events funded in 2024, using a binning graph
fig, ax = plt.subplots(figsize=(8,3))

x = df["Grant"]
y = df["Stream"]
ax.set_title("Cultural Festivals Funding by Program Stream, 2024",
             fontsize = 15)
ax.set_xlabel("Stream Type",
              fontsize = 12)
ax.set_ylabel("Number of Cultural Events",
              fontsize = 12)
ax.hist(y,
        color="green") 
```
