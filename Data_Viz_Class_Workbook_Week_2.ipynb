```python
# Import all libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
from PIL import Image
import requests
from io import BytesIO
```

```python
x = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110,180,240,99,220])
y2 = np.array([170,100,90,120,50])

fig, ax = plt.subplots(figsize=(5,3))

ax.bar(x,y1,color = "indigo")
ax.plot(x, y2, color = "red")
```

```python
y2_sd = np.std(y2)

fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x,y2, color="red")

ax.errorbar(x, y2,
            yerr = y2_sd,
            fmt = "none")
```

```python
y2_sd = np.std(y2)

fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x,y2, color="red")
#modifications to the error bar, color, line size and adding a capthick
ax.errorbar(x, y2,
            yerr = y2_sd,
            fmt = "none",
            ecolor = "green",
            elinewidth = 4,
            capsize = 6,
            capthick = 4,
            errorevery=2)
```

```python
response = requests.get('https://upload.wikimedia.org/wikipedia/en/c/cb/Monkey_D_Luffy.png')
image_file = BytesIO(response.content)
image = Image.open(image_file)
```

```python
fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x, y2, color ='pink', linestyle = "--")

ax_image = fig.add_axes([0.1, 0.11, 0.15, 0.35])

ax_image.imshow(image)
ax_image.axis("off")
```

```python
import seaborn as sns
import numpy as np
import pandas as pd
import PIL 
import requests 
import matplotlib.pyplot as plt
import scipy
# from wordcloud import wordCloud
# from matplotlib_venn import venn2, venn2_circles, venn2_unweighted
```

```python
tips = sns.load_dataset("tips")
print(tips)
```

```python
sns.lineplot(data = tips,
             x = 'total_bill',
             y = 'tip')
```

```python
tipgraph = sns.lineplot(data = tips,
                        x = 'total_bill',
                        y = 'tip')

tipgraph.set(title = 'Tips vs. Total Bill',
             xlabel = 'Total Bill ($)',
             ylabel = 'Tip Amount ($)')
```

```python
fig = plt.subplots(figsize=(10,5))

tipgraph = sns.lineplot(data = tips,
                        x = 'total_bill',
                        y = 'tip',
                        color = 'pink',
                        linestyle = '--',
                        linewidth = 3,
                        marker = 'o',
                        markerfacecolor = 'indigo')

tipgraph.set(title = 'Tips vs. Total Bill',
             xlabel = 'Total Bill ($)',
             ylabel = 'Tip Amount ($)')
```

```python
fig = plt.subplots(figsize=(10,5))

tipgraph = sns.scatterplot(data = tips,
                           x = 'total_bill',
                           y = 'tip',
                           style = 'time',
                           hue = 'day',
                           palette = ['purple', 'hotpink', 'deepskyblue', 'yellowgreen'])

tipgraph.set(title = 'Tip vs Total Bill',
             xlabel = 'Total Bill ($)',
             ylabel = 'Tip Amount ($)')
```

```python
sns.pairplot(data =tips,
             hue = 'day')
```

```python
daysplot = sns.relplot(
    data = tips,
    x='total_bill',
    y='tip',
    hue='sex',
    col='day',
    kind='scatter',
    col_wrap=2
)
daysplot
```

```python
import plotly.graph_objects as go 

x1 = np.array(['Liffu', 'Zoro', 'Nami', 'Usopp', 'Sanji'])
y1 = np.array([110,180,240,99,220])
```

```python
graph = go.Figure()

graph.add_trace(go.Bar(x = x1, y = y1))

graph.update_layout(title = "Pirate Scores",
                    xaxis_title = 'Pirates', 
                    yaxis_title = 'Score')
```

```python
graph = go.Figure()

graph.add_trace(go.Scatter(
                x=x1,
                y=y1,
                mode = 'markers',
                marker = dict(
                    size = 15,
                    color="hotpink",
                    opacity = 1,
                    line=dict(width = 5,
                              color = 'purple')
                )
                ))

graph.update_layout(title='Interactive Pirate Plot',
                    xaxis_title = "Pirates",
                    yaxis_title = "Scores",
                    width=500,
                    height=500)
```
