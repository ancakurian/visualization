```python
#use "transforms" to position text 
#figure = the container
#ax = the specific point added
fig, ax = plt.subplots()
ax.axis([0,10,0,10])

ax.text(1,5,". Data: (1,5)", transform=ax.transData)
ax.text(0.5,0.1,". Axes: (0.5, 0.1)", transform=ax.transAxes)
ax.text(0.2,0.2,". Figure: (0.2, 0.2)", transform=fig.transFigure)

```

```python
np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0,75,50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np.array([110,180,240,99,220])
```

```python
# coding subplots, with x and y, no of columns, no of rows

fig, (ax1, ax2) = plt.subplots(ncols=2,nrows=1,figsize=(7,3), layout="constrained")
plt.xlabel("ExPeRiMeNtInG with Labels", color = 'green', fontsize = 16)
ax1.scatter(x1,y1, marker='o',color='yellow')
ax2.bar(x2,y2, color='green')

plt.xticks(rotation=60,
           fontsize = 10)
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                    ['ax2', 'ax3']],
                                    figsize=(10,5),
                                    layout="constrained")
someaxes["ax1"].scatter(x1,y1, marker='*', color='red')
someaxes["ax2"].bar(x2,y2, color='yellow')
someaxes["ax3"].plot(x1,y1, linestyle='--')

someaxes["ax1"].set_xlabel("Big Label", color='red', fontsize=18, rotation=15)
someaxes["ax2"].set_xlabel("Another Label", color='yellow', fontsize=14, rotation=15)
someaxes["ax3"].set_xlabel("Label 2: MISSING", color='blue', fontsize=26)
```
